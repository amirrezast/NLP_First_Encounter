```python
import nltk
from nltk.corpus import stopwords, state_union, twitter_samples, movie_reviews
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.sentiment import SentimentIntensityAnalyzer
import random

nltk.download(["names",
               "stopwords",
               "state_union",
               "twitter_samples",
               "movie_reviews",
               "averaged_perceptron_tagger",
               "vader_lexicon",
               "punkt"])
```

```python
# Loading text data from the State of the Union corpus
sotu = state_union.raw()

# Tokenizing the text
words = word_tokenize(sotu)

# Filtering out non-alphabetic tokens and converting to lower case
words = [word.lower() for word in words if word.isalpha()]

# Removing stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

# Displaying the most common words
freq_dist = FreqDist(filtered_words)
print(freq_dist.most_common(10))
```

```python
# Tokenizing the words from the raw text
tokens = word_tokenize(sotu)
```

```python
# Frequency distribution
fdist = FreqDist(filtered_words)

# Displaying the most common words visually
fdist.tabulate(10)
```

```python
# Initializing the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Example sentiment analysis on a sample text
sample_text = "This is an excellent movie. I really enjoyed it!"
sentiment_scores = sia.polarity_scores(sample_text)
print(sentiment_scores)
```

```python
# Loading Twitter sample data
tweets = [tweet.replace("://", "//") for tweet in twitter_samples.strings('positive_tweets.json')]

# Shuffle tweets
random.shuffle(tweets)

# Getting sentiment scores for each tweet
for tweet in tweets:
    print(tweet, sia.polarity_scores(tweet))
```

```python
# Function to determine if a text is positive based on the VADER score
def is_positive(text):
    score = sia.polarity_scores(text)['compound']
    return score > 0.05

# Testing the function on movie reviews
positive_reviews = [review for review in movie_reviews.fileids('pos')]
negative_reviews = [review for review in movie_reviews.fileids('neg')]

# Calculating accuracy
correct = 0
for review in positive_reviews:
    text = movie_reviews.raw(review)
    if is_positive(text):
        correct += 1

for review in negative_reviews:
    text = movie_reviews.raw(review)
    if not is_positive(text):
        correct += 1

accuracy = correct / (len(positive_reviews) + len(negative_reviews))
print(f"Accuracy: {accuracy}")
```
